# Introduction to Bit Manipulation : AND, OR, XOR, NOT, Shift Operations

```python
def convert2bin(num:int) -> str:
    if num == 0:
        return "0"

    result = ""
    while num > 0:
        if num % 2 == 1:
            result += "1"
        else:
            result += "0"

        num = num//2
    
    result = result[::-1]
    return result

convert2bin(15)
```

```python
def convert2dec(x:str)->int:
    decimal_num = 0
    power = 0
    index = len(x)-1
    while index >= 0:
        num = int(x[index]) * (2**power)
        decimal_num += num
        index -= 1
        power += 1
    return decimal_num

# TC = O(n), SC = O(1)
convert2dec("1101")
```

```python
# 1's Complement
def ones_complement(binary:str) -> int:
    result = ""
    
    for bit in binary:
        if bit == "0":
            result += "1"
        else:
            result += "0"

    return result 

print(ones_complement("101101"))
```

```python
# 2's Complement
def twos_complement(binary:str) -> str:
    #Step 1
    ones = ""

    for bit in binary:
        if bit == "0":
            ones += "1"
        else:
            ones += "0"

    # Step 2
    carry = 1
    result = list(ones)

    for i in range(len(result) - 1, -1, -1):
        if carry == 0:
            break
        if result[i] == "0":
            result[i] = "1"
            carry = 0
        else:
            result[i] = "0"

    return "".join(result)

print(twos_complement("101101"))
```

```python
# AND
def binary_and(a: str, b: str) -> str:
    result = ""

    for i in range(len(a)):
        if a[i] == '1' and b[i] == '1':
            result += '1'
        else:
            result += '0'
    
    return result

binary_and("1101","0011")
```

```python
# OR
def binary_or(a: str, b: str) -> str:
    result = ""

    for i in range(len(a)):
        # if (a[i] == '1' and b[i] == '1') or (a[i] == '1' and b[i] == '0') or (a[i] == '0' and b[i] == '1'):
        if (a[i] == '1' or b[i] == '1'):
            result += '1'
        else:
            result += '0'
    
    return result

binary_or("1101","0001")
```

```python
# XOR
def binary_xor(a:str, b:str) -> str:
    result = ""

    for i in range(len(a)):
        if (a[i] == '1' and b[i] == '0') or (a[i] == '0' and b[i] == '1'):
            result += '1'
        else:
            result += '0'

    return result

binary_xor("1101","0001")
```

```python
# OR
def binary_not(a: str) -> str:
    result = ""

    for i in a:
        if i == '1':
            result += '0'
        else:
            result += '1'
    
    return result

binary_not("1101")
```

```python
# Left Shift 
def left_shift(a: str, shift: int) -> str:
    return a[shift:] + ('0' * shift)

def right_shift(a: str, shift: int) -> str:
    return ('0' * shift) + a[:-shift]

print("Left :", left_shift("1101", 1))
print("Right:", right_shift("1101", 1))
```
