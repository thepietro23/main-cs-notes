# Reverse Doubly Linked List

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.prev = None
        self.next = None

class DoublyLinkedlist:
    def __init__(self):
        self.head = None

    def insert_at_head(self, val):
        new_node = Node(val)
        if self.head is None:
            self.head = new_node

        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node

    def insert_at_between(self, val, position):
        new_node = Node(val)
        if position == 0:
            self.insert_at_head(val)
            return
        
        current = self.head
        count = 0
        while current and count < position - 1:
            current = current.next
            count += 1

        if current is None:
            print('Position is outbound')
            return

        new_node.next = current.next
        new_node.prev = current
        if current.next:
            current.next.prev = new_node
        current.next = new_node

    def append(self, val):
        new_node = Node(val)
        if self.head is None:
            self.head = new_node

        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
            new_node.prev = current

    def traverse(self):
        if self.head is None:
            print('DLL is empty')
            return
        current = self.head
        while current is not None:
            print(current.val, end=" <-> ")
            current = current.next
        print('None')

    def traverse_reverse(self):
        if self.head is None:
            print('DLL is empty')
            return
        current = self.head
        while current.next is not None:
            current = current.next
        while current is not None:
            print(current.val, end=" <-> ")
            current = current.next
        print('None')


    def reverse_dll(self):
        stack = []
        temp = self.head
        while temp is not None:
            stack.append(temp.val)
            temp = temp.next
        
        temp = self.head
        while temp is not None:
            e = stack.pop()
            temp.val = e
            temp = temp.next

        return temp

    def reverse_dll_oa(self):
        if self.head is None or self.head.next is None:
            return self.head

        current = self.head
        prev = None
        
        while current is not None:
            front = current.next
            current.next = prev
            current.prev = front

            prev = current
            current = front

        self.head = prev
        return self.head


        
```

# Delete All Occurrences of a Key in Doubly Linked List

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.prev = None
        self.next = None

class DLL:
    def __init__(self):
        self.head = None

    def insert_at_head(self, val):
        new_node = Node(val)
        if self.head is None:
            self.head = new_node
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node

    def insert_at_between(self, val, position):
        new_node = Node(val)
        if position == 0:
            self.insert_at_head(val)
            return

        current = self.head
        count = 0
        if current and count < position - 1:
            current = current.next
            count += 1

        if current is None:
            print("Position is outbound")
            return

        new_node.next = current.next
        new_node.prev = current
        if current.next:
            current.next.prev = new_node
        current.next = new_node

    def append(self, val):
        new_node = Node(val)
        if self.head is None:
            self.head = new_node

        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
            new_node.prev = current

    def traverse(self):
        if self.head is None:
            print('DLL is empty')
            return
        current = self.head
        while current is not None:
            print(current.val, end=" <-> ")
            current = current.next
        print('None')

    def dkidll(self, key):

        if self.head.next is None and self.head.val == k:
            return None
        temp = self.head
        prev = None
        new_head = self.head

        while temp is not None:
            if temp.val == key:
                if temp.prev is not None:
                    temp.prev.next = temp.next
                else:
                    self.head = temp.next

                if temp.next is not None:
                    temp.next.prev = temp.prev

                next_node = temp.next
                temp.prev = None
                temp.next = None
                temp = next_node
            else:
                temp = temp.next

    
    # Find Pairs with Given Sum in Doubly Linked List
    def fpdll(self, target):
        temp1 = self.head
        result = []

        while temp1 is not None:
            temp2 = temp1.next
            while temp2 is not None:
                if temp1.val + temp2.val == target:
                    result.append([temp1.val, temp2.val])
                temp2 = temp2.next

            temp1 = temp1.next

        return result

    #Better Solution
    def fpdll_bs(self, target):
        my_set = set()
        temp = self.head
        result = []

        while temp is not None:
            remaining = target - temp.val
            if remaining in my_set:
                result.append([remaining, temp.val])     
            my_set.add(temp.val) 
            temp = temp.next
            
        return result

    def fpdll_oa(self, target):
        left = self.head
        right =  self.head
        result = []

        while right.next is not None:
            right = right.next

        while left is not None and right is not None and left != right and  left.prev != right:
            total = left.val + right.val

            if total == target:
                result.append([left.val, right.val])
                left = left.next
                right = right.prev
            elif total > target:
                right = right.prev
            else:
                left = left.next
        return result

    # Remove Duplicates from a Sorted Doubly Linked List
    def rdll(self):
        if self.head is None:
            return self.head

        current = self.head

        while current is not None and current.next is not None:
            if current.val == current.next.val:
                dup = current.next
                current.next = dup.next
                if dup.next is not None:
                    dup.next.prev = current

                dup.prev = None
                dup.next = None
            else:
                current = current.next

        return self.head


        

```
