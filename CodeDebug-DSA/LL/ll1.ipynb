# Singly Linked List

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.next = None

class SinglyLinkedList:
    def __init__(self) -> None:
        self.head = None

    def append(self, val):
        new_node = Node(val)
        # SLL is empty
        # if self.head == None:       # empty list: new node becomes head
        if not self.head:
            self.head = new_node
        else:                       # non-empty: traverse to tail, then attach
            # SLL is not empty
            current = self.head
            while current.next is not None:
                current = current.next
            current.next = new_node
    # Traverse
    def traversal(self):
        if self.head is None:
            print('SLL is empty')
        else:
            current = self.head
            while current is not None:
                print(current.val, end=" ")
                current = current.next
            print()

    # Insert
    def insert(self, val, position):
        # if self.head is None:
        new_node = Node(val)
        # Insert at starting / Head
        if self.head is None or position == 0:
            new_node.next = self.head
            self.head = new_node
        # Insert in the middle
        else:
            count = 0
            prev_node  = None
            current = self.head
            while current is not None and count < position:
                prev_node = current
                current = current.next
                count += 1
            prev_node.next = new_node
            new_node.next = current

    def delete(self, val, position):
        temp = self.head
        if temp.next is not None:
            if temp.val == val:
                self.head = temp.next
                return
        # temp.next = None
        # del temp
            else:
                found = False
                prev_node = None
                count = 0
                while temp is not None:
                    if temp.val == val:
                        found = True
                        break
                    prev_node = temp
                    temp = temp.next

                if found:
                    prev_node.next = temp.next
                    # del temp
                    return
                else:
                    print("Node not found")

# TC - O(N)
# SC - O(1)


                
            
```

```python
sll = SinglyLinkedList()
sll.traversal()
```

```python
sll.append(10)
sll.traversal()
```

```python
sll.append(20)
sll.append(30)
sll.append(40)
sll.traversal()
```

```python
sll.insert(1,2)
sll.traversal()
```
