# Find the middle of a linked list | Tortoise-Hare Approach

```python
# Brute force

class Node:
    def __init__(self, val):
        self.val = val
        self.next = None

class SingleLL:
    def __init__(self):
        self.head = None

    def append(self, val):
        new_node = Node(val)
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next is not None:
                current = current.next
            current.next = new_node

    def traverse(self):
        if self.head is None:
            print("LL is empty")
        else:
            current = self.head
            while current is not None:
                print(current.val, end=" ")
                current = current.next
            print()

    # Brute Force
    def traverse_middle(self):
        n = 0
        temp = self.head
        while temp is not None:
            n += 1
            temp = temp.next
        temp = self.head
        for i in range(0, n//2):
            temp = temp.next
        return temp.val
    
    # Optimal Approach
    def tortoise_hare(self):
        slow = self.head
        fast = self.head
        while fast is not None and fast.next is not None:
            slow = slow.next
            fast = fast.next.next
        return slow.val

```

# Reverse a Linked List | Iterative Approach

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.next = None

class SLL:
    def __init__(self):
        self.head = None

    def append(self, val):
        new_node = Node(val)
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next is not None:
                current = current.next
            current.next = new_node
    
    def traverse(self):
        if self.head is None:
            print('LL is empty')
        else:
            current = self.head
            while current is not None:
                print(current.val, end=' ')
                current = current.next
            print()

    # Brute Force
    def reverse_ll(self):
        # Stak m value dalaenge from LL
        temp = self.head
        stack = []
        while temp is not None:
            stack.append(temp.val)
            temp=temp.next

        # Ab retrive krenege val from stck
        temp = self.head
        while temp is not None:
            e = stack.pop()
            temp.val = e
            temp = temp.next 
        return self.head

    # Optimal
    def rev_ll_optimal(self): 
        temp = self.head
        prev_node = None
        
        while temp is not None:
            front = temp.next
            temp = prev_node
            prev_node = temp
            temp = front

        return self.head
```

#  Linked List Cycle | Floyd’s Cycle Detection 

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.next = None

class SLL:
    def __init__(self):
        self.head = None

    def append(self, val):
        new_node = Node(val)

        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next is not None:
                current = current.next  
            current.next = new_node

    def traverse(self):
        if self.head is None:
            print("LL is empty")
        else:
            current  = self.head
            while current is not None:
                print(current.val, end=" ")
                current = current.next
            print()

    def ll_cycle(self):
        temp = self.head
        my_set = set()
        while temp is not None:
            if temp in my_set:
                return True
            my_set.add(temp)
            temp = temp.next

        return False

    def ll_cycle_optimal(self):
        slow = self.head
        fast = self.head

        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next

            if slow == fast:
                return True

        return False

    def start_ll_cycle(self):
        temp = self.head
        my_set = set()
        while temp is not None:
            if temp in my_set:
                return temp.val
            my_set.add(temp)
            temp = temp.next
        return None

    # Linked List Cycle II | Find the Cycle Starting Point
    def start_ll_cycle_optimal(self):
        slow = self.head
        fast = self.head
        while fast and fast.next is not None:
            slow = slow.next
            fast = fast.next.next

            if slow == fast:
                slow = self.head
                while slow != fast:
                    slow = slow.next
                    fast = fast.next

                return slow.val
        return None

    # Find Length of Loop in Linked List TC-O(N), SC-O(N)
    def len_of_loop_ll(self):
        temp = self.head
        travel=0
        my_dict = {}
        while temp is not None:
            if temp in my_dict:
                return travel - my_dict[temp]
            my_dict[temp]=travel
            travel += 1
            temp = temp.next

        return None

    def len_of_loop_ll_optimal(self):
        slow = self.head
        fast = self.head
        while fast is not None and fast.next is not None:
            slow = slow.next
            fast = fast.next.next
            if slow == fast:
                slow = slow.next
                count = 1
                while slow != fast:
                    slow=slow.next
                    count +=1 
                return count
        return count


```

# Leetcode 328: Odd Even Linked List | Rearrange Nodes 

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class SLL:
    def __init__(self):
        self.head = None

    def append(self, value):
        new_node = Node(value)

        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next is not None:
                current = current.next
            current.next = new_node

    def traverse(self):
        if self.head is None:
            print("SLL is empty")
        else:
            current = self.head
            while current is not None:
                print(current.value, end=" ")
                current = current.next
            print()



     # TC = O(n), SC = O(n)
    def odd_even_ll_bf(self):
        if self.head is None or self.head.next is None:
            # print("LL is empty for Odd-Even LL")
            return self.head

        temp = self.head
        values=[]

        while temp:
            values.append(temp.value)
            temp = temp.next.next if temp.next else None
        
        temp = self.head.next
        while temp:
            values.append(temp.value)
            temp = temp.next.next if temp.next else None

        index = 0
        temp = self.head
        while temp is not None:
            temp.value = values[index]
            index += 1
            temp = temp.next
        return self.head    

    # TC = O(n), SC = O(1)               
    def odd_even_ll_oa(self):
        if self.head is None or self.head.next is None:
            return self.head

        odd = self.head
        even = self.head.next
        even_head = even

        while even is not None and even.next is not None:
            odd.next = odd.next.next
            odd = odd.next
            even.next = even.next.next
            even = even.next

        odd.next = even_head

        return self.head

    # TC = o(n/2)
    # sc = o(1)

    #Leetcode 19: Remove Nth Node from End of List | Two Pointer Approach
    def rndfel_bf(self, n):
        length = 0 
        temp = self.head
        while temp is not None:
            length += 1
            temp = temp.next
        if length == n:
            new_head = self.head.next
            # del head
            return new_head

        position_to_stop = length - n
        temp = self.head
        count = 1
        while count < position_to_stop:
            temp = temp.next
            count += 1
        temp.next = temp.next.next
        return self.head

    def rndffel_oa(self, n):
        slow = self.head
        fast = self.head
        for _ in range(n):
            fast = fast.next

        if fast == None:
            return self.head.next
        
        while fast.next is not None:
            slow = slow.next
            fast = fast.next

        slow.next = slow.next.next
        return self.head


```
