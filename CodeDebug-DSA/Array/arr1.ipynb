### Print all factors of a given numbers

```python
# Brute force method to find factors of a number
# Time complexity: O(n)
# Space complexity: O(n)
def factor(num):
    result = []
    for i in range(1, num + 1):
        if num % i == 0:
            result.append(i)
    return result

num = 34
factor(num)
```

```python
# Optimized method to find factors of a number
# Time complexity: O(sqrt(n))
# Space complexity: O(n)

import math
def factor_oa(num):
    result = []
    for i in range(1, int(math.sqrt(num))+1):
        if num % i == 0:
            result.append(i)
            if i != num // i:
                result.append(num//i)
    return sorted(result)


num = 34
factor_oa(num)
```

## Store Frequency in Dictionary

```python
# Time complexity: O((n))
# Space complexity: O(1)
def frequency_map(arr):
    fm = {}
    for i in range(len(arr)):
        if arr[i] in fm:
            fm[arr[i]] += 1
        else:
            fm[arr[i]] = 1
            
    print(fm[1])
    return fm

arr = [5,6,7,7,1,9,11,1,1,5,1,1]
frequency_map(arr)
```

```python
# Time complexity: O((n))
# Space complexity: O(1)
def freq_map(arr):
    hm = {}
    n = len(arr)

    for i in range(0, n):
        hm[arr[i]] = hm.get(arr[i], 0) + 1

    return hm

arr = [5,6,7,7,1,9,11,1,1,5,1,1]
freq_map(arr)
```

```python
# Character Hashing

# Constraints
# 'a' <= s[i]<= 'z' >>> 97-122

def char_hashing(s):
    hash_list = [0]*26

    for i in s:
        index = ord(i) - ord('a')
        hash_list[index] += 1
    return hash_list
```

```python
s = "abbaca"
freq = char_hashing(s)

print(freq[ord('a') - ord('a')])  # count of 'a'
```

```python
# Brute Force

def first_non_repeating(s):
    n = len(s)

    for i in range(n):
        count = 0
        for j in range(n):
            if s[i] == s[j]:
               count += 1

        if count == 1:
            return s[i]

    return None     

s = "abbaca"
freq = first_non_repeating(s)

print(freq[ord('a') - ord('a')])  # count of 'a'
```

```python
def first_non_repeating_hash(s):
    freq = [0] * 26
    for ch in s:
        freq[ord(ch) - ord('a')] += 1

    for ch in s:
        if freq[ord(ch) - ord('a')] == 1:
            return ch
    
    return None

s = "abbaca"
print(first_non_repeating_hash(s))
```

```python
def first_non_repeating_ascii(s):
    freq = [0] * 128

    for ch in s:
        freq[ord(ch)] += 1

    for ch in s:
        if freq[ord(ch)] == 1:
            return ch

    return None


# Example
s = "abbaca"
print(first_non_repeating_ascii(s))
```

```python
def first_non_repeating_lowercase(s):
    freq = [0] * 26

    for ch in s:
        index = ord(ch) - ord('a')
        freq[index] += 1

    for ch in s:
        index = ord(ch) - ord('a')
        if freq[index] == 1:
            return ch

    return None


# Example
s = "abbaca"
print(first_non_repeating_lowercase(s))
```
