# 🧠 1️⃣ What is Head Recursion?

### 👉 Definition

When **recursive call happens first**, and work happens **after recursion returns**, it is called **Head Recursion**.

---

### Example

```python
def head(n):
    if n == 0:
        return
    head(n - 1)
    print(n)
```

Call: `head(3)`

Execution order:

* Go deep first
* Print during backtracking

Output:

```
1
2
3
```

---

### 🧠 Mental Model

```
Call stack builds completely
Then work happens while stack unwinds
```

---

# 🧠 2️⃣ What is Tail Recursion?

### 👉 Definition

When **recursive call is the last statement**, and nothing happens after it, it is called **Tail Recursion**.

---

### Example

```python
def tail(n):
    if n == 0:
        return
    print(n)
    tail(n - 1)
```

Call: `tail(3)`

Output:

```
3
2
1
```

---

### 🧠 Mental Model

```
Work happens before recursive call
Nothing remains after recursion
```

---

# 🔥 Key Difference

| Feature                 | Head Recursion  | Tail Recursion   |
| ----------------------- | --------------- | ---------------- |
| Recursive call position | First           | Last             |
| Work done               | After recursion | Before recursion |
| Stack unwinding needed? | Yes             | No logical need  |
| Can optimize to loop?   | Harder          | Easy             |

---

# 🧠 Time Complexity

Both examples:

Each function calls itself n times.

### ⏱ Time Complexity:

```
O(n)
```

For both head and tail recursion.

---

# 🧠 Space Complexity

Each recursive call creates stack frame.

Max depth = n

### 📦 Space Complexity:

```
O(n)
```

For both.

---

# 🔥 Advanced Interview Insight

### Tail Call Optimization (TCO)

Some languages (like C, C++) can optimize tail recursion into iteration.

So tail recursion can become:

```
O(n) time
O(1) space
```

BUT ⚠️

👉 Python does NOT support tail call optimization.

So in Python:

```
Space Complexity = O(n)
```

Even for tail recursion.

Here’s your crisp revision sheet:

---

## 📌 Head vs Tail Recursion

**Head Recursion**

* Recursive call happens first.
* Work is done during stack unwinding.
* Cannot be easily converted to loop.
* Example: printing numbers in ascending order.

**Tail Recursion**

* Recursive call is last statement.
* No work after recursive call.
* Can be optimized to iteration (in some languages).
* More memory efficient if TCO supported.

---

## ⏱ Time Complexity

Both:

```
O(n)
```

## 📦 Space Complexity

Without TCO:

```
O(n)
```

With TCO:

```
O(1)
```

---

# 🚀 CP Insight

In competitive programming:

❌ Avoid recursion if depth > 10^5
✔ Prefer iteration for large constraints
✔ Use tail recursion only if depth small

What is Recursion Using Parameters?

Instead of:
 >> Returning values and combining results later

We:
1. Pass values as parameters
2. Accumulate answer during recursive calls
3. This often converts head recursion → tail recursion.

````python
# Classic - Head Recursion
def print_1_to_n(n):
    if n == 0:
        return
    print_1_to_n(n-1)
    print(n)

print_1_to_n(3)
````

````python
# Using Parameter

def print_1_to_n_p(i,n):
    if n == 0:
        return 
    print(i)
    print_1_to_n_p(i, n-1)

print_1_to_n_p(15,5)
````

````python
# Print 1 to N using Tail Recursion 
def print_1_to_n(i, n):
    if i > n:
        return
    print(i)
    print_1_to_n(i + 1, n)

print_1_to_n(1, 5)
````

#  Factorial with recursion

````python
#  Factorial with recursion

def fact(n):
    if n == 0 or n == 1:
        return 1
    return n * fact(n-1)

fact(5)
````

# Reverse an array using recursion

````python
# Reverse an array using recursion

def reverse_rec(left, right, arr):
    # left = arr[0]
    # right = arr[len(arr)] - 1
    if left >= right:
        return
    arr[left], arr[right] = arr[right], arr[left]
    reverse_rec(left+1, right-1, arr)
    # print(result)

arr = [1,2,3,4,5,6,7,8,9]
reverse_rec(1, 4, arr)
print(arr)
````

````python
# If a string is an palindrome
def str_rec(left, right, arr):

    while left < right:
        if arr[left] != arr[right]:
            return False
        left += 1
        right -= 1
    return True
        

arr = ['a','a','b','b','a','a']
arr1 = 'aabbaa'
str_rec(0,5,arr1)
````

````python
# TC - O(N)
def rec_rev_str(left, right, arr):
    if left >= right:
        return True
    if arr[left] != arr[right]:
        return False
    
    rec_rev_str(left+1, right-1, arr)

arr1 = 'aabbaa'
str_rec(0,5,arr1)
````

# Fibonacci Number

````python
def fib_rec(num):
    if num <= 1:
         return num
  
    return fib_rec(num-1) + fib_rec(num-2)
    
num = 7
fib_rec(num)
````

# Better — Memoization (Top-Down DP)

````python
def fib_memo(n, dp):
    if n <= 1:
        return n
    if dp[n] != -1:
        return dp[n]
    
    dp[n] = fib_memo(n-1, dp) + fib_memo(n-2, dp)
    return dp[n]

n = 7
dp = [-1] * (n+1)
fib_memo(n, dp)
````
