# Largest Element in an array

```python
def largest_element(arr):
    
    max_element = arr[0]
    for i in range(len(arr)):
        if arr[i] >= max_element:
            max_element = arr[i]
            i += 1
    
    return max_element
 

arr = [55,32,-97,99,3,67]
largest_element(arr)
```

```python
def le_2(arr):
    largest = arr[0]
    for i in range(len(arr)):
        largest = max(largest, arr[i])

    return largest

le_2(arr)
```

# 2nd Largest Element without sort

```python
def sec_le(arr):
    if len(arr) < 2:
        return -1
    
    largest = float('-inf')
    sec_largest = float('-inf')
    for i in range(len(arr)):
        largest = max(largest, arr[i])
        if arr[i] != largest and arr[i] > sec_largest:
            sec_largest = arr[i]

    return sec_largest if sec_largest != -float('inf') else -1

arr = [55,32,-97,99,3,67]
sec_le(arr)
```

```python
def sec_le_2(arr):
    if len(arr) < 2:
        return -1
    
    largest = float('-inf')
    sec_largest = float('-inf')

    for i in range(len(arr)):
        if arr[i] > largest:
            sec_largest = largest
            largest = arr[i]
        elif arr[i] > sec_largest and sec_largest !=  largest:
            sec_largest = arr[i]

    return sec_largest if sec_largest != -float('inf') else 1

sec_le_2(arr)
```

# Check if array is sorted or not 

```python
def check_arr_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(i+1, n):
            if arr[i] > arr[j]:
                return False
            else:
                i+=1

        return True
    
arr1 = [1,2,3,4,5,6]
arr2 = [4,5,6,2,1]
print(check_arr_sort(arr1))
print(check_arr_sort(arr2))
```

```python
def cas(arr):
    n = len(arr)
    for i in range(0, n):
        if arr[i] > arr[i+1]:
            return False
    return True

arr1 = [1,2,3,4,5,6]
arr2 = [4,5,6,2,1]
print(check_arr_sort(arr1))
print(check_arr_sort(arr2))
```

# Remove Duplicates from SOrted Array

```python
# TC = O(2N)
# SC = O(N)

def rdfs(arr):
    n = len(arr)
    freq_map = {}
    for i in range(n):
        freq_map[arr[i]] = 0
    j = 0
    for k in freq_map:
        arr[j] = k
        j += 1

    return j


arr = [1,1,1,2,3,4,5,7,9,9,9,10]
rdfs(arr) #gives total unique number
```

```python
# Optimal - two pointer

def rdfs(arr):
    n = len(arr)
    if n == 1:
        return 1
    
    i = 0
    j = i+1

    while j < n:
        if arr[j] != arr[i]:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
        j += 1

    return i+1

arr = [1,1,1,2,3,4,5,7,9,9,9,10]
rdfs(arr)
```

# Right Rotate Array by 1 Place

```python
def rrab1p(arr):
    n = len(arr)
    # inplace\
    res = [0]*n
    res[0] = arr[n-1]

    for i in range(n-1):
        res[i+1] = arr[i]

    return res


arr = [5,-2,3,9,0,6,10,7]
rrab1p(arr)
```

```python
def rrab1p_oa(arr):
    n = len(arr)
    temp = arr[n-1]

    for i in range(n-2,-1,-1):
        arr[i+1] = arr[i]
    arr[0] = temp
    return arr

arr = [5,-2,3,9,0,6,10,7]
rrab1p_oa(arr)
```

# Right Rotate Array by K Place

```python
# Brute force
def rrabkp(arr, k):
    n = len(arr)
    rotations = k%n
    for _ in range(rotations):
        e = arr.pop()
        arr.insert(0,e)

    return arr


arr = [3,9,5,6,7,2]
rrabkp([3,9,5,6,7,2],4)
```

```python
# slicing

def rrabkp2(arr, k):
    n = len(arr)
    rotation = k%n
    arr[:] = arr[n-rotation:] + arr[:n-rotation]
    return arr

rrabkp2([3,9,5,6,7,2],4)
```

```python
# Optimal

def reverse(arr, left, right):
    while left < right:
        arr[left], arr[right] = arr[right], arr[left]
        left += 1
        right -= 1

def rrabkp_oa(arr, k):
    n = len(arr)
    k = k%n
    reverse(arr, n-k,n-1)    # Reverse Last K element
    reverse(arr, 0, n-k-1)  #Reverse remaining elements
    reverse(arr, 0, n-1)    #Reverse whole array

    return arr

rrabkp_oa([3,1,5,6,7,2,10,9],5)
```

# Move zeros to the end


```python
# Btute force
# TC- O(N)
# SC- O(N)

def mzte1(arr):
    n1 = len(arr)
    temp = []
    for i in range(0,n1):
        if arr[i] != 0:
            temp.append(arr[i])
    n2 = len(temp)
    for i in range(n2):
        arr[i] = temp[i]

    for i in range(n2, n1):
        arr[i] = 0

    return arr

arr = [1,0,2,4,3,0,0,3,5,1]
mzte1(arr)
```

```python
# Optimal Approach - Single iteration
# TC- O(N)
# SC- O(1)
def mzte2(arr):
    n = len(arr)
    if n == 1:
        return 
    
    i = 0
    while i < n:
        if arr[i] == 0:
                break
        i += 1

    if i == n:
            return
        
    j = i+1

    while j < n:
        if arr[j] != 0:
            arr[i], arr[j] = arr[j], arr[i]
            i+=1
        j+= 1
    return arr

arr = [1,0,2,4,3,0,0,3,5,1]
mzte2(arr)
```

# Linear Search

```python
def linear_search(arr, target):
    n = len(arr)

    for i in range(n):
        if target == arr[i]:
            return i
        
    return -1
        
arr = [2,1,4,6,8,3,5]
linear_search(arr, 5)
```
