```python
# BS

def bs1(arr, target):
    low, high = 0, len(arr)-1
    while low <= high:
        mid = low + (high-low)//2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid+1
        else:
            high = mid-1
    return -1

arr = [2,4,6,7,9,11,18,19,20]
target = 7
bs1(arr, target)
```

```python
# Recursive

def bs(arr, low, high):
    mid = low + (high-low)//2
    if low > high:
        return -1
    if arr[mid] == target:
        return mid
    elif arr[mid] < target:
        return bs(arr, mid+1, high)
    else:
        return bs(arr, low, mid)
    
arr = [2,4,6,7,9,11,18,19,20]
target = 7
bs(arr, 0, len(arr)-1)

    
```

# Lower Bound

```python
# Smallest Index such that nums[i] >= target
# TC: O(logn)
# SC: O(1)
def lb(arr, target):
    n = len(arr)
    lb = n
    low, high = 0, n-1
    while low<= high:
        mid = low + (high-low)//2
        if arr[mid] >= target:
            lb = mid
            high = mid - 1
        else:
            low = mid + 1
    return arr[lb]


arr = [1,1,1,2,3,3,5,6,7,7,7,9,12,12,13]
target = 7
lb(arr, target)
```

# CEIL THE FLOOR

```python
# CEIL THE FLOOR
def ctf(arr, target):
    n = len(arr)
    low, high = 0, n-1
    floor, ceil = None, None
    while low <= high:
        mid = low + (high-low)//2
        if arr[mid] == target:
            return arr[mid], arr[mid]
        elif arr[mid] > target:
            ceil = arr[mid]
            high = mid - 1
        else:
            floor = arr[mid]
            low = mid + 1

    return f"Target: {target} | Ceil: {ceil}  Floor: {floor}"


arr = [3,4,4,4,8,9,9,10,12,12,14,15]
target = 16
ctf(arr, target)
```

# Count Occurrences in Sorted Array

```python
# Count the occurrence of a target in a sorted array
def countOcc(arr, target):
    n = len(arr)
    first = -1
    last = -1
    low, high = 0, n-1
    for i in range(n):
        if arr[i] == target:
            if first == -1:
                first = i
            last = i
    return [first, last, last-first+1]

arr = [1,2,3,3,3,3,3,5,6,8,9,9,10]
target = 3
countOcc(arr, target)
```

# Search In Rotated Sorted Array

```python
#  Search in rotated sorted array

# Brute Force: O(n)

def search_rotated(arr, target):
    n = len(arr)
    for i in range(n):
        if arr[i] == target:
            return i
    return -1
    
arr = [1,4,5,6,8,10,11,15,20]
target = 13
search_rotated(arr, target)
```

```python
# Optimal Solution: O(log n)
def search_rotated_optimal(arr, target):
    n = len(arr)
    low, high = 0, n-1
    while low <= high:
        mid = low + (high-low)//2
        if arr[mid] == target:
            return mid
        elif arr[mid] >= arr[low]:
            if arr[low] <= target <= arr[mid]:
                high = mid-1
            else:
                low = mid + 1
        else:
            if arr[mid] <= target <= arr[high]:
                low = mid + 1
            else:
                high = mid - 1
    return -1

arr = [12,6,8,10,11,15,20,1,4,5]
target = 10
search_rotated_optimal(arr, target)
```

```python
# Search in Rotated Sorted Array II (Duplicates allowed)
def search_rotated_duplicates(arr, target):
    n = len(arr)
    low, high = 0, n-1
    while low <= high:
        mid = low + (high-low)//2
        if arr[mid] == target:
            return True
        if arr[low] == arr[mid] == arr[high]:
            low += 1
            high -= 1
            continue

        if arr[mid] <= arr[high]:
            if arr[mid] < target <= arr[high]:
                low = mid + 1
            else:
                high = mid - 1
        else:
            if arr[low] <= target < arr[mid]:
                high = mid - 1
            else:
                low = mid + 1
    return False


arr = [7,7,7,7,7,7,7,1,2,3,4,5,7,7]
target = 1
search_rotated_duplicates(arr, target)
```

```python
def min_rotated_sorted(arr):
    n = len(arr)
    mini = float('inf')
    for i in range(n):
        mini = min(mini, arr[i])

    return mini

arr = [4,5,6,7,0,1,2]
min_rotated_sorted(arr)
```
